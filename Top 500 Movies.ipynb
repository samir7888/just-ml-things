# TOP 500 Movies


```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('Top 500 Movies Ranked by Combined Critics and Audience Scores.csv')
```

```python
#what is the datatypes of the each columns?
df.sample(4)
```

```python
#is there any missing values?

df.isnull().sum()
```

```python
# how columns looks mathmatically ?
df.describe()
```

```python
# Are there any duplicate values ?
df.duplicated().sum()
```

```python
#Categorical data

# sns.countplot(df['Year'])
df['Year'].value_counts().iloc[:5].plot(kind = 'pie', autopct = '%.2f')
```

```python
# Numerical data

#Histogram(create range acts like categorical, same like distribution)

import matplotlib.pyplot as plt
plt.hist(df['IMDb_10'],bins = 20)
```

```python
#to find the probability of occurance of particular imdb rating we use distplot(Distrubution Plot)

# sns.distplot(df['IMDb_10'])
sns.histplot(data=df, x="IMDb_10", kde=True)
```

```python
# We use Boxplot to find potential outlieres and can remove them

sns.boxplot(df['IMDb_10'])
```
