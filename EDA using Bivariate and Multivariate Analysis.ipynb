```python
import seaborn as sns
import pandas as pd
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
flights = sns.load_dataset('flights')
```

```python
# ScatterPlot (Numerical - Numerical)
sns.scatterplot(data = tips, x = tips['total_bill'], y= tips['tip'], hue = tips['sex'], style=tips['smoker'], size= tips['size'])
```

```python
# BarPLot (Numerical - Categorical)
sns.barplot(data = tips , x = tips['smoker'], y = tips['total_bill'] , hue= tips['sex'])
```

```python
# BoxPlot (Numerrical - Categorical
sns.boxplot(data = tips , x = tips['smoker'], y = tips['total_bill'] , hue= tips['sex'])
```

```python
#Distplot( Numerical - Categorical) - Distribution

sns.distplot(tips[tips['sex']=='Male']['total_bill'], hist='false')
sns.distplot(tips[tips['sex']=='Female']['total_bill'], hist='false')

```

```python
sns.kdeplot(data=tips, x='total_bill', hue='sex')
# sns.kdeplot(data=tips[tips['sex'] == 'Male'], x='total_bill')
# sns.kdeplot(data=tips[tips['sex'] == 'Female'], x='total_bill')
```

```python
#HeatMap ( Categorical - categorical)

sns.heatmap(pd.crosstab(tips['sex'], tips['Size']))
```
